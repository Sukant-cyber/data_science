```python
plant=["Spinach","Turnip","Rhubarb","Brocoli","Kale"]
died=[10,25,5,30,21]
germinated=[74,88,56,59,59]
plt.figure()
plt.bar(plant,died,color="#6fa8dc",bottom=germinated,width=0.4, label='Died Seedlings' )
plt.bar(plant,germinated,color="#6aa84f",width=0.4,label='Germinated Seedlings')
plt.title("Seeds Germinated",fontsize=14)
plt.ylabel("Number of seeds")
plt.legend(frameon=True,facecolor='white',shadow=True)

```

```python
import pandas as pd
titanic=pd.read_csv('data/titanic.csv')
```

```python
titanic.info()
```

```python
titanic['age']=pd.to_numeric(titanic['age'],errors='coerce')
```

```python
# titanic['age']
```

```python
plt.hist(titanic['age'])
```

```python
plt.hist(titanic['age'],bins=10)
```

```python
first_class=titanic[titanic['pclass']==1]
```

```python
plt.hist(first_class['age'])
```

```python
second_class=titanic[titanic['pclass']==2]
```

```python
plt.hist(second_class['age'])
```

```python
third_class=titanic[titanic['pclass']==3]
```

```python
plt.hist(third_class['age'])
```

```python
first_class=titanic[titanic['pclass']==1]['age']
second_class=titanic[titanic['pclass']==2]
third_class=titanic[titanic['pclass']==3]
plt.hist(first_class,label="First_class",alpha=0.5)
plt.hist(second_class['age'],label="Second_class",alpha=0.5)
plt.hist(third_class['age'],label="Third_class",alpha=0.5)
plt.legend(frameon=True,shadow=True,facecolor='white')
```

```python
nums=np.arange(5)
plt.figure(figsize=(9,4),dpi=120)
plt.suptitle("Overall Title",fontsize=24)
plt.subplot(1,3,1)
plt.plot(nums,nums,label='linear')
plt.legend(loc='upper left')
plt.title('Linear')
plt.subplot(1,3,2)
plt.plot(nums,nums**2,color="red",label='square')
plt.legend(loc='upper left')
plt.title('Square')
plt.subplot(1,3,3)
plt.plot(nums,num**3,color="orchid",label='cube')
plt.legend(loc='upper left')
plt.title('Cube')
plt.tight_layout()
plt.show()
```
