```python
import matplotlib.pyplot as plt
```

```python
# Chart 1 - Total Vulnerabilities by Rating 

ratings = ['Critical', 'High', 'Medium', 'Low']
values = [1, 9, 3, 1]

fig, ax = plt.subplots(figsize=(6, 3))

# Swap axes
ax.bar(ratings, values, width=0.5, color='#dddddd') 

ax.set_xticks(range(len(values)))
ax.set_xticklabels(ratings)

for i, v in enumerate(values):
    ax.text(i, v + 0.1, str(v), ha='center')
    
ax.set_ylabel('Number of Vulnerabilities')
# ax.set_title('Total Vulnerabilities by Rating', fontsize=10)

plt.tight_layout()
```

```python
# Chart 2 - Average by Risk Factor

factors = ['Threat Agents', 'Vulnerability Factors', 'Technical Impact', 'Business Impact']

# Calculate averages 
threat_agents = (8 + 6 + 5 + 1) / 4
vuln_factors = (8 + 8 + 1 + 1) / 4
tech_impact = (8 + 7 + 5 + 2) / 4  
biz_impact = (8 + 7 + 5 + 1) / 4

values = [threat_agents, vuln_factors, tech_impact, biz_impact]

# Plot chart
fig, ax = plt.subplots(figsize=(6, 3))
ax.barh(factors, values, height=0.5, color='#dddddd')

# Remaining chart code
```

```python
# Chart 3 - Average Overall Rating

values = [4.25]
labels = ['Overall']

fig, ax = plt.subplots(figsize=(6, 3))

# Use light blue color 
ax.pie(values, labels=labels, autopct='%.2f', colors=['lightblue'], wedgeprops={'width': 1})

ax.set_title('Average Overall Rating', fontsize=10)

plt.tight_layout()
```
